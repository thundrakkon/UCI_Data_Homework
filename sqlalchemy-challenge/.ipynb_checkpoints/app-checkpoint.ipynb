# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# Query All Records in the the Database
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
connection = engine.connect()
```

```python
# reflect an existing database into a new model
# Declare bases and map class
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Determine the last date in the data
session.query(Measurement.date).order_by(Measurement.date.desc()).first()
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database
year_ago = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print(year_ago)
```

```python
# Perform a query to retrieve the date and precipitation scores
date_prec = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date >= year_ago).\
    order_by(Measurement.date).all()
date_prec
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the dataframe by date
df = pd.DataFrame(date_prec, columns=['date', 'prcp'])
df.set_index('date', inplace=True)
df
```

```python
# Use Pandas Plotting with Matplotlib to plot the data

y_axis = df["prcp"]
x_axis = df.index.values

# Change figure size and set figure to 1
plt.figure(figsize = [9, 5])
plt.figure(1)

# Plot bar graph
plt.bar(x_axis,
        y_axis,
        width=2.0)

plt.tight_layout()

# Set ticks labels, rotate 90 degrees, and change font size
plt.xticks(['2016-08-23',
            '2016-10-01',
            '2016-11-01',
            '2016-12-01',
            '2017-01-01',
            '2017-02-01',
            '2017-03-01',
            '2017-04-01',
            '2017-05-01',
            '2017-06-01',
            '2017-07-01',
            '2017-08-01',
            '2017-08-23'],
           rotation = 90,
           fontsize = 10
          )

#plt.grid(False)

# Set title
plt.title("Hawaii Precipation",
          fontsize = 20,
          fontweight = "bold",
          color = "darkblue")

# Set y-axis label
plt.ylabel("Rainfall (inches)",
           fontsize = 14,
           color = "darkblue")

# Set x-axis label
plt.xlabel("Dates",
           fontsize = 14,
           color = "darkblue")

# Show plot
plt.show()
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
sel = [Measurement.station, 
       func.count(Measurement.id)]

station_counts = session.query(*sel).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.id).desc()).all()
station_counts
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?

active_station = session.query(Measurement.station).\
    group_by(Measurement.station).\
    order_by(func.count(Measurement.id).desc()).first()

# Lowest temperature recorded
min_temp = session.query(func.min(Measurement.tobs)).filter(Measurement.station == 'USC00519281').first()

# Highest temperature recorded
max_temp = session.query(func.max(Measurement.tobs)).filter(Measurement.station == 'USC00519281').first()

# Average temperature
avg_temp = session.query(func.avg(Measurement.tobs)).filter(Measurement.station == 'USC00519281').first()

print(f"The lowest temperature recorded for station {active_station} is {min_temp} degrees F")
print(f"The highest temperature recorded for station {active_station} is {max_temp} degrees F")
print(f"The average temperature for station {active_station} is {avg_temp} degrees F")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

temp_most = session.query(Measurement.tobs).\
    filter(Measurement.date >= year_ago).\
    filter(Measurement.station == 'USC00519281').\
    order_by(Measurement.tobs).all()
temp_most
```

```python
# Create a dataframe
df2 = pd.DataFrame(temp_most, columns=['tobs'])
df2.set_index('tobs', inplace=True)
df2
```

```python
# Plot the results as a histogram

# Set x values
x = df2.index.values

# Change figure size and set figure to 2
plt.figure(figsize = [9, 5])
plt.figure(2)

# Plot histogram
bins = 12
plt.hist(x, bins, alpha=0.5)
plt.tight_layout()

#plt.grid(False)

# Set title
plt.title("Hawaii's Most Frequent Station Temperature",
          fontsize = 20,
          fontweight = "bold",
          color = "darkblue")

# Set y-axis label
plt.ylabel("Frequency",
           fontsize = 14,
           color = "darkblue")

# Set x-axis label
plt.xlabel("Temperature",
           fontsize = 14,
           color = "darkblue")

# Show plot
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
