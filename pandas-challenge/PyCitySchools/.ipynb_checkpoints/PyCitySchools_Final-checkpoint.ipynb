```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(school_data_to_load)
student_df = pd.read_csv(student_data_to_load)

student_df.loc[student_df['reading_score']>=70 , 'read_pass'] = 1
student_df.loc[student_df['math_score']>=70 , 'math_pass'] = 1

# Combine the data into a single dataset
school_dset = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
school_dset.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
schools_total = len(school_dset["school_name"].unique())

#Calculate the total number of students
students_total = school_dset["student_name"].count()

#Calculate the total budget
school_group = school_dset.groupby(['school_name'])
budget_school = school_group['budget'].unique()
budget_total = budget_school.sum()

#Calculate the average math score
math_avg = school_dset['math_score'].mean()

#Calculate the average reading score
read_avg = school_dset['reading_score'].mean()

#Calculate the percentage of students with a passing math score (70 or greater)
pass_math = school_dset['math_pass'].count()/students_total

#Calculate the percentage of students with a passing math score (70 or greater)
#pass_read = (school_dset.loc[school_dset['reading_score'] >= 70, 'reading_score'].count())/students_total
pass_read = school_dset['read_pass'].count()/students_total

#Calculate the overall passing rate (overall average score)
pass_rate = (pass_math + pass_read)/2
```

```python
#Create a dataframe to hold the above results
district_summary = pd.DataFrame({'Total Schools': [schools_total],
                                'Total Students': [students_total],
                                'Total Budget': budget_total,
                                'Average Math Score': math_avg,
                                'Average Reading Score': read_avg,
                                '%Passing Math': pass_math,
                                '%Passing Reading': pass_read,
                                '%Overall Passing Rate': pass_rate})

#Optional: give the displayed data cleaner formatting
district_format = {'Total Students': '{:,}',
                  'Total Budget': '${:,.2f}',
                  '%Passing Math': '{:.2%}',
                  '%Passing Reading': '{:.2%}',
                  '%Overall Passing Rate': '{:.2%}'}

district_summary.style.format(district_format).hide_index()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Group by School Name
group_school = school_dset.groupby(['school_name'])

#School Type
school_type = group_school['type'].unique()

#Total Students
total_students = group_school['student_name'].count()

#Total School Budget
school_budget = group_school['budget'].unique()

#Per Student Budget
student_budget = school_budget/total_students

#Average Math Score
math_school = group_school['math_score'].mean()

#Average Reading Score
read_school = group_school['reading_score'].mean()

#% Passing Math
student_pass_math = group_school['math_pass'].count()/total_students

#% Passing Reading
student_pass_read = group_school['read_pass'].count()/total_students

#Overall Passing Rate (Average of the above two)
overall_pass = (student_pass_math + student_pass_read)/2
```

```python
#Create a dataframe to hold the above results
school_summary = pd.DataFrame({'School Type': school_type,
                               'Total Students': total_students,
                               'Total School Budget': school_budget,
                               'Per Student Budget': student_budget,
                               'Average Math Score': math_school,
                               'Average Reading Score': read_school,
                               '%Passing Math': student_pass_math,
                               '%Passing Reading': student_pass_read,
                               '%Overall Passing Rate': overall_pass})

school_summary['School Type'] = school_summary['School Type'].str[0]
school_summary['Total Students'] = school_summary['Total Students'].map('{:,}'.format)
school_summary['Total School Budget'] = school_summary['Total School Budget'].astype(float).map('${:,.2f}'.format)
school_summary['Per Student Budget'] = school_summary['Per Student Budget'].astype(float).map('${:,.2f}'.format)
school_summary['%Passing Math'] = school_summary['%Passing Math'].astype(float).map('{:.2%}'.format)
school_summary['%Passing Reading'] = school_summary['%Passing Reading'].astype(float).map('{:.2%}'.format)
school_summary['%Overall Passing Rate'] = school_summary['%Overall Passing Rate'].astype(float).map('{:.2%}'.format)

school_summary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
grades = school_dset['grade'].sort_values(ascending = True).unique()
num_grades = len(grades)

for x in range(0, num_grades):
    school_dset_x = school_dset.loc[school_dset["grade"] == grades[x], :]
    math_df = pd.DataFrame({'School Name': school_dset_x['school_name'],
                            'Math': school_dset_x['math_score']})
    group_x = math_df.groupby(['School Name'])
    avg_math = group_x['Math'].mean()
    avg_df_x = pd.DataFrame({grades[x]: avg_math})
    
    if x == 0:
        school_grade_math = avg_df_x
    elif x > 0:
        school_grade_math = pd.merge(school_grade_math, avg_df_x, on=["School Name"], how="left")

school_grade_math
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
grades = school_dset['grade'].sort_values(ascending = True).unique()
num_grades = len(grades)

for x in range(0, num_grades):
    school_dset_x = school_dset.loc[school_dset["grade"] == grades[x], :]
    read_df = pd.DataFrame({'School Name': school_dset_x['school_name'],
                            'Reading': school_dset_x['reading_score']})
    group_x = read_df.groupby(['School Name'])
    avg_read = group_x['Reading'].mean()
    avg_df_x = pd.DataFrame({grades[x]: avg_read})
    
    if x == 0:
        school_grade_read = avg_df_x
    elif x > 0:
        school_grade_read = pd.merge(school_grade_read, avg_df_x, on=["School Name"], how="left")

school_grade_read
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#Group by School Name
group_school = school_dset.groupby(['school_name'])

#School Type
school_type = group_school['type'].unique()

#Total Students
total_students = group_school['student_name'].count()

#Total School Budget
school_budget = group_school['budget'].unique()

#Per Student Budget
student_budget = school_budget/total_students

#Average Math Score
math_school = group_school['math_score'].mean()

#Average Reading Score
read_school = group_school['reading_score'].mean()

#% Passing Math
student_pass_math = group_school['math_pass'].count()/total_students

#% Passing Reading
student_pass_read = group_school['read_pass'].count()/total_students

#Overall Passing Rate (Average of the above two)
overall_pass = (student_pass_math + student_pass_read)/2
```

```python
new_school_summary = pd.DataFrame({'Total Students': total_students,
                                   'School Type': school_type,
                                   'Total School Budget': school_budget,
                                   'Per Student Budget': student_budget,
                                   'Average Math Score': math_school,
                                   'Average Reading Score': read_school,
                                   '%Passing Math': student_pass_math,
                                   '%Passing Reading': student_pass_read,
                                   '%Overall Passing Rate': overall_pass})

new_school_summary['Spending Ranges (Per Student)'] = pd.cut(new_school_summary['Per Student Budget'], bins, labels=group_names)

new_spending = new_school_summary.groupby(['Spending Ranges (Per Student)'])

new_math_avg = new_spending['Average Math Score'].mean()
new_read_avg = new_spending['Average Reading Score'].mean()
new_math_pct = new_spending['%Passing Math'].mean()
new_read_pct = new_spending['%Passing Reading'].mean()
new_overall_pct = new_spending['%Overall Passing Rate'].mean()

spending_scores = pd.DataFrame({'Average Math Score': new_math_avg,
                                'Average Reading Score': new_read_avg,
                                '%Passing Math': new_math_pct,
                                '%Passing Reading': new_read_pct,
                                '%Overall Passing Rate': new_overall_pct})

spending_scores['%Passing Math'] = spending_scores['%Passing Math'].astype(float).map('{:.2%}'.format)
spending_scores['%Passing Reading'] = spending_scores['%Passing Reading'].astype(float).map('{:.2%}'.format)
spending_scores['%Overall Passing Rate'] = spending_scores['%Overall Passing Rate'].astype(float).map('{:.2%}'.format)


spending_scores
```
